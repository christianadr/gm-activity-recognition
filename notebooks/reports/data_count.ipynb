```python
import matplotlib.pyplot as plt
%matplotlib inline
from pathlib import Path

from constants import globals as g
```

```python
data_dir = Path(g.RAW_DIR, "gm-activities")
```

```python
data_count = dict()

for subdir in list(data_dir.iterdir()):
    data_count[subdir.name] = len(list(subdir.iterdir()))

data_count
```

```python
data_count.keys()
```

```python
plt.figure(figsize=(10, 5))
bars = plt.bar(data_count.keys(), height=data_count.values())

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(height), 
            ha='center', va='bottom', fontsize=8)

plt.title("Data count: No. of clips per skill")
plt.xlabel('GM Skills')
plt.ylabel('Number of Clips')

# textstr = 'Each GM Skill must have 100 video clips before\ntreating it with data preprocessing and training\nusing PoseC3D.'
# props = dict(boxstyle='round,pad=1', facecolor='lightgray', edgecolor='black')

# plt.text(4.25, 60, textstr, fontsize=6, verticalalignment='top', bbox=props)

plt.show()
```
